```
import pandas as pd
import numpy
from matplotlib import pyplot as plt

import numpy as np
```

```
excel_file = 'y.xlsx'
tocode = pd.read_excel(excel_file)
# opening Excel File
tocode
```

```
Average_Availability_SaaS_ID = tocode["Availability"].mean(axis=0)

#Average_Availability_SaaS_ID


# calculating Average of Availability of SaaSID
```

```
tocode['SP(Availability)Sensitivity ^^2'] = tocode['Availability']**2


#calculating SP(Availability)Sensitivity ^^2
```

```
tocode['SLA * SP(Availability)Sensitivity ^^2'] = tocode['SP(Availability)Sensitivity ^^2']*0.99

#calculating SLA * SP(Availability)Sensitivity ^^2
```

```
tocode['SLA* SP(Availability)Sensitivity ^^2 / Average Availability of SaaS_ID'] = tocode['SLA * SP(Availability)Sensitivity ^^2']/Average_Availability_SaaS_ID

#calculating SLA* SP(Availability)Sensitivity ^^2 / Average Availability of SaaS_ID
```

```
tocode['Percentage of Availability'] = numpy.floor(((tocode['SLA* SP(Availability)Sensitivity ^^2 / Average Availability of SaaS_ID']/0.99)*100))

#calculating Percentage for Availability table
```

```
availability_table = tocode[["SaaS_ID","Availability","SP(Availability)Sensitivity ^^2","SLA * SP(Availability)Sensitivity ^^2","SLA* SP(Availability)Sensitivity ^^2 / Average Availability of SaaS_ID","Percentage of Availability"]]
availability_table

#Total Availability Table Calculated and printed 
```

```
tocode['SP(Response Time)  Sensitivity ^^2'] = tocode['Response Time(ms)']**2


#calculating SP(Response Time)  Sensitivity ^^2
```

```
Average_Response_Time_of_SaaS_ID = tocode["Response Time(ms)"].mean(axis=0)


#calculating Average_Response_Time_of_SaaS_ID
```

```
tocode['SLA* SP(Response Time)  Sensitivity ^^2'] = tocode['SP(Response Time)  Sensitivity ^^2']*0.2

#calculating SLA* SP(Response Time)  Sensitivity ^^2
```

```
tocode['SLA* SP(Response Time)  Sensitivity ^^2 / Average Response Time of SaaS_ID'] = tocode['SLA* SP(Response Time)  Sensitivity ^^2']/Average_Response_Time_of_SaaS_ID


#calculating SLA* SP(Response Time)  Sensitivity ^^2 / Average Response Time of SaaS_ID
```

```
tocode['Percentage of Response time'] = (tocode['SLA* SP(Response Time)  Sensitivity ^^2 / Average Response Time of SaaS_ID']/0.2)*100


#calculating Percentage of Response time
```

```
response_table = tocode[["SaaS_ID","Response Time(ms)","SP(Response Time)  Sensitivity ^^2","SLA* SP(Response Time)  Sensitivity ^^2","SLA* SP(Response Time)  Sensitivity ^^2 / Average Response Time of SaaS_ID","Percentage of Response time"]]

#Total response Table Calculated and printed 

response_table
```

```
tocode['SP(Cost)  Sensitivity ^^2'] = tocode['Cost($)']**2
#calculating SP(Cost)  Sensitivity ^^2
```

```
Average_Cost_of_SaaS_ID = tocode["Cost($)"].mean(axis=0)


#calculating Average_Cost_of_SaaS_ID
```

```
tocode['SLA* SP(Cost)  Sensitivity ^^2'] = tocode['SP(Cost)  Sensitivity ^^2']*0.5

#calculating SLA* SP(Response Time)  Sensitivity ^^2
```

```
tocode['SLA* SP(Cost)  Sensitivity ^^2 / Average Cost of SaaS_ID'] = tocode['SLA* SP(Cost)  Sensitivity ^^2']/Average_Cost_of_SaaS_ID


#calculating SLA* SP(Response Time)  Sensitivity ^^2 / Average Response Time of SaaS_ID
```

```
Cost_table = tocode[["SaaS_ID","Cost($)","SP(Cost)  Sensitivity ^^2","SLA* SP(Cost)  Sensitivity ^^2","SLA* SP(Cost)  Sensitivity ^^2 / Average Cost of SaaS_ID"]]

#Total response Table Calculated and printed 

Cost_table
```

```
barWidth = 0.5
fig = plt.subplots(figsize =(12, 8)) 
   
# set height of bar 
toplot = tocode["SLA* SP(Availability)Sensitivity ^^2 / Average Availability of SaaS_ID"] 


X = np.arange(len(tocode["SaaS_ID"]))

   
# Set position of bar on X axis 
br1 = np.arange(len(toplot)) 
br2 = [x + barWidth for x in br1] 
br3 = [x + barWidth for x in br2] 
   
# Make the plot 

plt.bar(br2, toplot, color ='#ad815a', width = barWidth, 
        edgecolor ='black', label ='Availability') 

plt.title("Auditor SLA Assessment  for Availability")
# Adding Xticks  
plt.xlabel('', fontweight ='bold') 
plt.ylabel('', fontweight ='bold') 
plt.xticks([r + barWidth for r in range(len(toplot))], 
           tocode["SaaS_ID"])
plt.ylim([0.9875, 0.9905])
 
plt.yticks(np.arange(0.9875,0.9905, 0.0005))

plt.legend()
plt.show()
```
