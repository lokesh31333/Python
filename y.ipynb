```
import pandas as pd
import numpy
```

```
excel_file = 'y.xlsx'
tocode = pd.read_excel(excel_file)
# opening Excel File
tocode
```

```
Average_Availability_SaaS_ID = tocode["Availability"].mean(axis=0)

#Average_Availability_SaaS_ID


# calculating Average of Availability of SaaSID
```

```
tocode['SP(Availability)Sensitivity ^^2'] = tocode['Availability']**2


#calculating SP(Availability)Sensitivity ^^2
```

```
tocode['SLA * SP(Availability)Sensitivity ^^2'] = tocode['SP(Availability)Sensitivity ^^2']*0.99

#calculating SLA * SP(Availability)Sensitivity ^^2
```

```
tocode['SLA* SP(Availability)Sensitivity ^^2 / Average Availability of SaaS_ID'] = tocode['SLA * SP(Availability)Sensitivity ^^2']/Average_Availability_SaaS_ID

#calculating SLA* SP(Availability)Sensitivity ^^2 / Average Availability of SaaS_ID
```

```
tocode['Percentage of Availability'] = numpy.floor(((tocode['SLA* SP(Availability)Sensitivity ^^2 / Average Availability of SaaS_ID']/0.99)*100))

#calculating Percentage for Availability table
```

```
availability_table = tocode[["SaaS_ID","Availability","SP(Availability)Sensitivity ^^2","SLA * SP(Availability)Sensitivity ^^2","SLA* SP(Availability)Sensitivity ^^2 / Average Availability of SaaS_ID","Percentage of Availability"]]
availability_table

#Total Availability Table Calculated and printed 
```
